```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
%matplotlib inline
```

```python
# Zillow Median Home Values per Sq Ft ($) for Metro Areas in the US
med_metro = pd.read_csv('https://raw.githubusercontent.com/ajgarber20/Garber-Arvati_Final-Project/master/Home_Value_Data/Metro_MedianValuePerSqft_AllHomes.csv', encoding = 'latin-1')
med_metroT=med_metro.T
med_metro = med_metro.set_index(['RegionName'])
med_metroC = med_metro.drop(['RegionID'], axis=1)
med_metroC = med_metroC.drop(['SizeRank'], axis=1)
med_metroCT=med_metroC.T
med_metroCT
fig, ax = plt.subplots()
ax.set_ylabel('Value per Sq. Ft ($)')
ax.set_xlabel('Date')
plot = med_metroCT.plot(figsize=(20,20), title ='Median Home Value per Sq. Ft by Metro Area Over Time' )
plot.legend(loc='best')
```
